```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory

train_dataset = image_dataset_from_directory("C:/Users/kavet/Onedrive/Documents/DeerGuardTraining/train", image_size=(640, 640), batch_size=661)
val_dataset = image_dataset_from_directory("C:/Users/kavet/Onedrive/Documents/DeerGuardTraining/valid", image_size=(640, 640), batch_size=661)
```

```python
model = tf.keras.applications.MobileNetV2(input_shape=(640, 640, 3), weights=None, classes=2)
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(train_dataset, validation_data=val_dataset, epochs=10)
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

with open("model.tflite", "wb") as f:
    f.write(tflite_model)
```
